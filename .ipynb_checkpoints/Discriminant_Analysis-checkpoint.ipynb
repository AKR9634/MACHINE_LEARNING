```python
X = scaled_dataset[['flavanoids', 'proline']].values
y = scaled_dataset['target'].values
```

### LINEAR DISCRIMINANT ANALYSIS

```python
mean0 = np.mean(X[y==0], axis = 0)
mean1 = np.mean(X[y==1], axis = 0)
mean2 = np.mean(X[y==2], axis = 0)

prior0 = np.mean(y==0)
prior1 = np.mean(y==1)
prior2 = np.mean(y==2)
```

```python
cov0 = np.cov(X[y==0].T)
cov1 = np.cov(X[y==1].T)
cov2 = np.cov(X[y==2].T)
```

```python
weighted_cov = (cov0 * (len(X[y == 0]) - 1) + cov1 * (len(X[y == 1]) - 1) + cov2 * (len(X[y == 2]) - 1)) / (len(X) - 3)  
```

```python
cov_inverse = np.linalg.inv(weighted_cov)
```

```python
cov_inverse
```

```python
#Now we will define the discriminant function for the linear discriminant analysis

def disc(x, mean, prior):
    return x @ cov_inverse @ mean -0.5 * mean.T @ cov_inverse @ mean + np.log(prior)

def predict(x):
    d0 = np.array([disc(z, mean0, prior0) for z in x])
    d1 = np.array([disc(z, mean1, prior1) for z in x])
    d2 = np.array([disc(z, mean2, prior2) for z in x])

    return np.argmax(np.array([d0, d1, d2]), axis = 0)
```

### DECISION BOUNDARY FOR LDA

```python
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
grid = np.c_[xx.ravel(), yy.ravel()]
```

```python
z = predict(grid).reshape(xx.shape)
```

```python
fig, ax = plt.subplots(figsize = (13, 13))
ax.contourf(xx, yy, z, alpha = 0.3, cmap = plt.cm.Blues)
ax.scatter(X[:, 0], X[:, 1], c = y, cmap = plt.cm.Blues, edgecolors = 'k', marker = 'o')
```

### QUADRATIC DISCRIMINANT ANALYSIS 

```python
classes = np.unique(y)
n = len(classes)

means = []
covariances = []
priors = []
inv_covs = []
det_covs = []

for c in classes: 
    X_c = X[y==c]
    means.append(np.mean(X_c, axis = 0))
    cov = np.cov(X_c.T)
    covariances.append(cov)
    priors.append(len(X_c)/ len(X))
    inv_covs.append(np.linalg.inv(cov))
    det_covs.append(np.linalg.det(cov))
    

means = np.array(means)
covariances = np.array(covariances)
priors = np.array(priors)
inv_covs = np.array(inv_covs)
det_covs = np.array(det_covs)

def QDA_disc(x, c):
    return (-0.5 * np.log(det_covs[c])) + (-0.5 * (x-means[c]).T @ inv_covs[c] @ (x-means[c])) + np.log(priors[c])


def predict_QDA(x):
    preds = []
    for z in x:
        scores = [QDA_disc(z,c) for c in classes]
        preds.append(np.argmax(scores))
    return np.array(preds)
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))

grid_points = np.c_[xx.ravel(), yy.ravel()]
```

```python
z = predict_QDA(grid_points).reshape(xx.shape)
```

### DECISION BOUNDARY FOR QDA

```python
fig, ax = plt.subplots(1, figsize = (13, 13))
ax.contourf(xx, yy, z, alpha = 0.3, cmap = plt.cm.Blues)
ax.scatter(X[:, 0], X[:, 1], c = y, cmap = plt.cm.Blues, edgecolors = 'k', marker = 'o')
```
