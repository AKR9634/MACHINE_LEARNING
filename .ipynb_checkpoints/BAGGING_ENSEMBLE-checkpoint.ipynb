```python
df_bag = df_train.sample(8, replace = True)
X = df_bag.iloc[:, 0:2]
y = df_bag.iloc[:, -1]
df_bag
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
```

```python
def evaluate(clf, X, y):
    clf.fit(X, y)
    plot_tree(clf)
    plt.show()
    plot_decision_regions(X.values, y.values, clf = clf, legend= 2)
    y_pred = clf.predict(X_test)
    print(accuracy_score(y_test, y_pred))
```

```python
dt_bag1 = DecisionTreeClassifier()
```

```python
evaluate(dt_bag1, X, y)
```

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X, y = make_classification(n_samples=10000, n_features=10, n_informative=3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
dt = DecisionTreeClassifier(random_state=3)
dt.fit(X_train, y_train)
print("Decision Tree Accuracy: ", accuracy_score(y_test, dt.predict(X_test)))
```

```python
#BAGGING
```

```python
bag = BaggingClassifier(estimator = DecisionTreeClassifier(), n_estimators=500, max_samples=0.25, bootstrap=True, random_state=3)
```

```python
bag.fit(X_train, y_train)
```

```python
accuracy_score(y_test, bag.predict(X_test))
```

```python
bag.estimators_samples_
```

```python
bag.estimators_samples_[0].shape
```

```python
#BAGGING USING SVM
```

```python
bag = BaggingClassifier(estimator = SVC(), n_estimators=500, max_samples=0.25, bootstrap=True, random_state=3)
```

```python
bag.fit(X_train, y_train)
```

```python
print("BAGGING USING SVM: ", accuracy_score(y_test, bag.predict(X_test)))
```

```python
#PASTING (for pasting ,just change the parameter Bootstrap from true to false)
```

```python
#RANDOM SUBSPACES
```

```python
bag = BaggingClassifier(estimator = DecisionTreeClassifier(), n_estimators=500, max_samples=1.0, bootstrap=False, max_features=0.5, bootstrap_features=True, random_state=3)
```

```python
bag.fit(X_train, y_train)
print("RANDOM SUBSPACES CLASSIFIER: ", accuracy_score(y_test, bag.predict(X_test)))
```

```python
#RANDOM PATCHES(for random patches just use row and column sampling together using max_samples and max_featrues parameter in the  bagging classifier.)
```

```python
#OOB SCORE(OUT OF BAG SAMPLING, rows which are not used in the algorithm due to sampling approx. 37%)
```

```python
#We can use to check the performance of the model
```

```python
bag = BaggingClassifier(estimator = DecisionTreeClassifier(), n_estimators=500, max_samples=0.25, bootstrap=True, oob_score = True, random_state=3)
```

```python
bag.fit(X_train, y_train)
```

```python
bag.oob_score_
```

```python
accuracy_score(y_test, bag.predict(X_test))
```
