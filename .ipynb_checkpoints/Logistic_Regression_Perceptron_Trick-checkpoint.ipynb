```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=20)
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y,cmap='winter')
```

```python
#Using step function
```

```python
def step(s):
    if(s<=0):
        return 0
    return 1
```

```python
def perceptron(X, y):

    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(10000):
        j = np.random.randint(0, X.shape[1])
        weights = weights + lr * (y[j] - step(np.dot(X[j], weights))) * X[j] 

    return weights[0], weights[1:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 3)
```

```python
intercept_, coef_ = perceptron(X_train, y_train)
```

```python
coef_
```

```python
intercept_
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'winter')
x_input = np.linspace(-3, 3, 100)
y_input = (-coef_[0]/coef_[1]) * x_input + (-intercept_/coef_[1]) 
plt.plot(x_input, y_input, color = 'red')
plt.ylim(-3, 2)
```

```python
def perceptron(X, y):

    m = []
    b = []

    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.01
    
    for i in range(10000):
        j = np.random.randint(0, X.shape[1])
        weights = weights + lr * (y[j] - step(np.dot(X[j], weights))) * X[j] 
        m.append(-weights[1]/weights[2])
        b.append(-weights[0]/weights[2])

        
    return m, b
```

```python
m, b = perceptron(X_train, y_train)
```

```python
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize = (9, 5))

x_i = np.arange(-3, 3, 0.1)
y_i = x_i * m[0] + b[0]

ax.scatter(X[:, 0], X[:, 1], c=y, cmap = 'winter', s=100)

line,  = ax.plot(x_i, x_i*m[0]+b[0], 'r-', linewidth = 2)
plt.ylim(-3, 3)

def update(i):
    label = 'epoch {0}'.format(i+1)
    line.set_ydata(x_i*m[i]+b[i])
    ax.set_xlabel(label)

anim = FuncAnimation(fig, update, repeat = True, frames = 200, interval = 100)
```

```python
#Logistic regression implementation using sigmoic function
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def perceptron(X, y):
    
    X = np.insert(X, 0, 1, axis = 1)
    lr = 0.1
    weights = np.ones(X.shape[1])
    
    for i in range(1000):
        arb = np.random.randint(0, X.shape[1])
        weights = weights + lr * (y[arb] - sigmoid(np.dot(X[arb], weights))) * X[arb]

    return weights[0], weights[1:]
```

```python
intercept_, coef_ = perceptron(X_train, y_train)
```

```python
intercept_
```

```python
coef_
```

```python
m = -(coef_[0]/coef_[1])
b = -intercept_/coef_[1]
```

```python
x_input2 = np.linspace(-3, 3, 100)
y_input2 = m*x_input2+b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color = 'red', linewidth = 3)
plt.plot(x_input1, y_input1, color = 'black', linewidth = 3)
plt.plot(x_input2, y_input2, color = 'blue', linewidth = 3)
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'winter', s = 100)
plt.ylim(-3, 2)
```
