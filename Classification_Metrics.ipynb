```python
from sklearn.datasets import load_breast_cancer, load_iris, load_linnerud
import pandas as pd
import numpy as np
```

```python
df = load_breast_cancer(as_frame=True)
```

```python
df
```

```python
X = pd.DataFrame( df.data, columns=df.feature_names)
```

```python
X["State"] = df.target
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X.iloc[:, 0:-1], X.iloc[:, -1], test_size=0.2, random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score
```

```python
#Checking Accuracy using sklearn class
print("The Accuracy of Logistic Regression is: ", accuracy_score(y_test, clf1.predict(X_test)))
print("The Accuracy of Decision Tree is: ", accuracy_score(y_test, clf2.predict(X_test)))
```

```python
#Using Confusion Metric
confusion_matrix(y_test, clf1.predict(X_test))
```

```python
#Complete report of all classification metrics
from sklearn.metrics import classification_report
print(classification_report(y_test, clf1.predict(X_test)))
```

```python
print("For Logistic Regression Model: ")
print("-"*37)
print(pd.DataFrame(confusion_matrix(y_test, clf1.predict(X_test)), columns=list(range(0, 2))))
print("Precison Score: ", precision_score(y_test, clf1.predict(X_test)))
print("Recall Score: ", recall_score(y_test, clf1.predict(X_test)))
print("F1 Score: ", f1_score(y_test, clf1.predict(X_test)))
```
