```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from sklearn.datasets import make_blobs
```

```python
X, _= make_blobs(n_samples=300, centers = 3, n_features = 2, random_state= 43)
```

```python
N, D = X.shape
```

```python
K = 3
```

```python
np.random.seed(43)
```

```python
#Initialize parameters
mu = X[np.random.choice(N, K, replace = False)] #intial means
pi = np.full(K, 1/K) #equal mixing coefficients
sigma = [np.eye(D) for _ in range(K)] #identity covariances
```

```python
def gaussian(x, mean, cov):
    return multivariate_normal.pdf(x, mean = mean, cov = cov)
```

```python
#EM ALGORITHM
max_iter = 100

for step in range(max_iter):
    # E-step : compute responsibilities
    gamma = np.zeros((N, K))
    for k in range(K):
        gamma[:, k] = pi[k] * gaussian(X, mu[k], sigma[k])

    #Normalize over all clusters to get probabilities
    gamma /= gamma.sum(axis = 1, keepdims = True)

    # M-step : update parameters
    N_k = gamma.sum(axis = 0)

    for k in range(K):
        # Update means

        mu[k] = (1/N_k[k]) * np.sum(gamma[:, k][:, np.newaxis] * X, axis = 0)

        # Update covariances

        x_centered = X - mu[k]

        sigma[k] = (gamma[:, k][:, np.newaxis] * x_centered).T @ x_centered / N_k[k]

        # Update mixing coefficients

        pi[k] = N_k[k]/N
```

```python
# Plot Results

responsibilities = gamma
cluster_assignments = np.argmax(responsibilities, axis = 1)

plt.figure(figsize = (8, 6))

for k in range(K):
    plt.scatter(X[cluster_assignments == k, 0], X[cluster_assignments == k, 1], label = f"Cluster {k}")

plt.title("Manual GMM Clustering (EM Algorithm)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
