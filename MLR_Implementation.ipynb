```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=2)
```

```python
test_data = pd.concat([pd.DataFrame(X_test, columns = diab.feature_names), pd.DataFrame(y_test, columns = ['Diab_Progression'])], axis = 1)
```

```python
test_data
```

```python
class Mul_LR:
    def __init__(self):
        self.m = None;
        self.b = None;
    def fit(self, X_train, y_train):
        X_train = np.insert(X_train, 0, 1, axis = 1)
        betas = np.linalg.inv(np.dot(X_train.T, X_train)).dot(X_train.T).dot(y_train)
        self.m = betas[1:]
        self.b = betas[0]
    
    def predict(self, X_test):
        return np.dot(X_test, self.m)+self.b
        
```

```python
# SKLEARN MULTIPLE LINEAR REGRESSION
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test)
```

```python
mlr = Mul_LR()
```
