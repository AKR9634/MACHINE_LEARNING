```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with open('drugs.html', 'r') as f:
    res = f.read()
```

```python
soup = BeautifulSoup(res, 'lxml')
```

```python
soup.prettify()
```

```python
x = soup.find_all('div', class_ = 'card-body')
x.pop(0)
```

```python
name = []
price = []
amount = []

for i in x:
    name.append(i.find('a', 'text-inherit text-decoration-none').text.strip())
    price.append(i.find('strong').text.strip())
    amount.append(i.find('a', 'text-decoration-none text-muted').text.strip())
```

```python
df = pd.DataFrame({'Name' : name, 'Amount' : amount, 'Price' : price})
```
