```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=300, n_features = 1, n_informative= 1, n_targets = 1, noise = 50)
```

```python
X, y
```

```python
df = pd.DataFrame({'feature1': X[:, 0], 'target':y})
```

```python
df
```

```python
plt.scatter(df['feature1'], df['target'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_test
```

```python
test_data = pd.concat([pd.Series(X_test[:, 0]), pd.Series(y_test)], axis = 1)
```

```python
test_data
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test)
X[: , 0]
```

```python
pd.concat([pd.Series(X_test[:, 0]), pd.Series(lr.predict(X_test))], axis = 1)
```

```python
plt.scatter(X, y)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0][0].scatter(X, y)
axs[0][1].scatter(X, y)
axs[0][1].plot(X[:, 0], lr.predict(X), color = 'r')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('MAE: ', mean_absolute_error(y_test, lr.predict(X_test)))
```

```python
print('R2 Score: ', r2_score(y_test, lr.predict(X_test)))
```
